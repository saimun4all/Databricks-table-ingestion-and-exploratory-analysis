Row Filtering

```python
# Filter rows where release year is between 2000 and 2010
df_2000_2010 = df.filter((df.release_year >= 2000) & (df.release_year <= 2010))
display(df_2000_2010)
```

```python
# Filter rows where release year is between 2000 and 2010

from pyspark.sql.functions import col

df_filtered = df.filter(
    (col("release_year") >=2000) & (col("release_year") <= 2010)
)
df_filtered.show()
```

```python
df_filtered = df.filter(col("release_year").between(2000, 2010))
display(df_filtered)
```

```python
# Show all the movies from Marvel
df_marvel = df.filter(df.studio == "Marvel Studios")
display(df_marvel)
```

```python
# Show distinct movie industry

unique_industries = df.select("industry").distinct()
display(unique_industries)
```

```python
# Unique languages
unique_languages = df.select("language").distinct()
display(unique_languages)
```

```python
# Add a column in the table showing profit from the budget and revenue columns

df = df.withColumn("profit", col("revenue") - col("budget"))
display(df)
```

```python
# Rename a column such as revenue to total revenue
df = df.withColumnRenamed("revenue", "total_revenue")
df.printSchema()
```
